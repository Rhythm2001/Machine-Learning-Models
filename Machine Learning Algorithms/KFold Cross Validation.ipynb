```python
def get_score(model,X_train, X_test, Y_train, Y_test):
  model.fit(X_train, Y_train)
  return model.score(X_test, Y_test)
```

```python
from sklearn.model_selection import StratifiedKFold
sk = StratifiedKFold(n_splits= 3)
L_score= []
SVC_score= []
RF_score= []
for train_index, test_index in sk.split(iris.data, iris.target):
  X_train, X_test, Y_train, Y_test = iris.data[train_index], iris.data[test_index], iris.target[train_index], iris.target[test_index]
  L_score.append(get_score(LogisticRegression(max_iter=500, solver='liblinear' ), X_train, X_test, Y_train, Y_test))
  SVC_score.append(get_score(SVC(gamma='auto'), X_train, X_test, Y_train, Y_test))
  RF_score.append(get_score(RandomForestClassifier(n_estimators=50), X_train, X_test, Y_train, Y_test))
                                                                                      

```

```python
print(list(L_score))
def aver(s):
  return sum(s)/ len(s)
aver(L_score)

```

```python
print(list(SVC_score))
aver(SVC_score)
```

```python
print(list(RF_score))
aver(RF_score)
```

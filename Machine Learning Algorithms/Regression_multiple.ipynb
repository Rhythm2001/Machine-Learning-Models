```python
df = pd.read_csv("rates.csv")
df
```

```python
median_bedroom = int(df.bedroom.median())
```

```python
df.bedroom = df.bedroom.fillna(median_bedroom)
df
```

```python
regr = linear_model.LinearRegression()
regr.fit(df[['area', 'bedroom', 'age']], df.price)
```

```python
regr.intercept_
```

```python
regr.coef_
```

```python
regr.predict([[3333,3,17]])
```

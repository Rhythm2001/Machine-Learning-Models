```python
df = pd.read_csv("homeprices.csv")
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies], axis='columns')
merged
```

```python
final_data = merged.drop(['town','robinsville'], axis='columns')
final_data
```

```python
from sklearn import linear_model 
r = linear_model.LinearRegression()

```

```python
final = final_data.drop(['price'], axis='columns')
final
```

```python
Y = final_data.price
Y
```

```python
r.fit(final,Y)
```

```python
r.predict([[3333,0,0]])
```

```python
r.score(final,Y)*100
```

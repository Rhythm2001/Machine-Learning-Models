```python
df = pd.read_csv("spam.csv")
df
```

```python
df.groupby("Category").describe()
```

```python
df['spam_mail']= df['Category'].apply(lambda x:1 if x=='spam' else 0)
df
```

```python
drop_df= df.drop(['Category'], axis='columns')
drop_df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(drop_df.Message, drop_df.spam_mail, test_size=0.25)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
X_train_vec= vec.fit_transform(X_train)
X_train_vec.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
naive_model = MultinomialNB()
naive_model.fit(X_train_vec, Y_train)
```

```python
emails = ['Hey, lets catch up a movie tonight', 'Register for Pragati 2022']
email_convert = vec.transform(emails)
naive_model.predict(email_convert)
```

```python
X_test_vec = vec.transform(X_test)
X_test_vec.toarray()[:4]
naive_model.score(X_train_vec, Y_train)
print(f"The accuracy of the model is {naive_model.score(X_train_vec, Y_train)}")
```
